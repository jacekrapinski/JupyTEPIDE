```python
import snappy

class bcolors:
    HEADER = '\033[1m'
    BLUE = '\033[93m'
    ENDC = '\033[0m'


file = '/eodata/Landsat-5/TM/L1T/2011/11/16/LS05_RMPS_TM__GTC_1P_20111116T100218_20111116T100238_147386_0194_0039_166E/LS05_RMPS_TM__GTC_1P_20111116T100218_20111116T100238_147386_0194_0039_166E.TIFF/LT51940392011320ESA00_MTL.txt'
file = '/eodata/Landsat-7/ETM/L1GT/2017/01/15/LE70900762017015ASA00/LE70900762017015ASA00_MTL.txt'
file = '/eodata/Landsat-8/OLI_TIRS/L1GT/2017/08/09/LC80522132017221MTI00/LC80522132017221MTI00_MTL.txt'

p = snappy.ProductIO.readProduct(file)

print(bcolors.HEADER+"BANDS"+bcolors.ENDC)
for i in p.getBandNames():
    print(i)

md = p.getMetadataRoot()
for i in md.getElements():
    elementy = i.getElements()
    for el in elementy:
        print('\n'+bcolors.HEADER+el.getName()+bcolors.ENDC)
        for z in el.getAttributes():
            print(bcolors.BLUE+z.getName()+bcolors.ENDC, z.getData())
```
